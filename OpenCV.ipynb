```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread("/Users/soumyaranjanmohanty/Desktop/Soumya/IMG_20160709_163554895_HDR.jpg")
plt.imshow(img)
```

## Convert the image from BGR to RGB

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

## GRAY Scale Image


```python
image_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap= 'gray')
```

```python
# Plot the three channels of the image

fig, axes = plt.subplots(nrows =1, ncols= 3, figsize= (20,20))

for i in range(0,3):
    ax = axes[i]
    ax.imshow(img_rgb[:,:,i], cmap = 'gray')
plt.show()    
```

```python
resized_image = cv2.resize(img_rgb,(650,500))
cv2.imshow("window",resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resized_image = cv2.resize(img_rgb,(650,500))
cv2.rectangle(resized_image, pt1 = (100,150), pt2 = (200,300), color =(0,255,0), thickness = 3)
cv2.putText(img_rgb,"Hi",(5,10),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)
cv2.imshow("window",resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
