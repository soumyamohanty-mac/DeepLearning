```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x1 = [1,2,3,4,5]
y1 = [10,20,17,40,35]

x2 = [5,7,10,14,16]
y2 = [17,32,25,40,45]

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('plot graph')

plt.plot(x1,y1,'r',label = 'line 1')
plt.plot(x2,y2,'g' ,label = 'line 2')
plt.legend()
plt.show()
```
